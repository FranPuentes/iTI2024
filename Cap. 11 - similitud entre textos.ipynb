```python
from collections import Counter
import re

def text_to_bag_of_words(text):
    # Convertir todo el texto a minúsculas
    text = text.lower()

    # Eliminar puntuación y caracteres especiales utilizando expresiones regulares
    text = re.sub(r'[^\w\s]', ' ', text)

    # Dividir el texto en palabras
    words = text.split()

    # Contar la ocurrencia de cada palabra utilizando Counter de la biblioteca collections
    bag_of_words = Counter(words)

    return bag_of_words

def jaccard_distance(dict1, dict2):
    keys1 = set(dict1.keys())
    keys2 = set(dict2.keys())
    intersection = len(keys1.intersection(keys2))
    union = len(keys1) + len(keys2) - intersection
    return 1 - intersection / union
```

```python
import matplotlib.pyplot as plt

# Ejemplo de uso
text1 = """
El Titanic, esa majestuosa obra de la ingeniería naval, partió de Southampton el 10 de abril de 1912 con un aura de invencibilidad.
Construido con un despliegue tecnológico sin precedentes y un lujo sin parangón, su travesía transatlántica estaba destinada a entrar en los anales de la historia.
Sin embargo, en lugar de recibir elogios por su exitoso viaje, el Titanic quedó inmortalizado por una tragedia que nadie había anticipado.
Golpeando un iceberg la noche del 14 de abril de 1912, su "invulnerable" casco se rasgó, lo que llevó a que más de 1,500 personas perdieran la vida en las gélidas aguas del Atlántico Norte.
A pesar de la confianza en sus sistemas de seguridad, no había suficientes botes salvavidas para todos a bordo, lo que exacerbó la magnitud de la catástrofe.
"""
bag_of_words = text_to_bag_of_words(text1)
bow1=bag_of_words

for word in bag_of_words.keys():
    print(f"{word}: aparece {bag_of_words[word]} veces en el texto")

# Filtrar palabras con frecuencia mayor que 1
filtered_words = {word: freq for word, freq in bag_of_words.items() if freq > 1}

# Ordenar las palabras por frecuencia
sorted_words = sorted(filtered_words.items(), key=lambda x: x[1], reverse=True)

# Extraer las palabras y sus frecuencias en listas separadas
words, frequencies = zip(*sorted_words)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies, color='blue', width=1.6)  # Ancho de la barra establecido en 1.6
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.title('Bag-of-Words')
plt.xticks(rotation=90, fontsize=12)  # Hacer que las etiquetas del eje x aparezcan en vertical
plt.yticks(fontsize=10)
for i, freq in enumerate(frequencies):
    plt.text(i, freq, str(freq), ha='center', va='bottom')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Ejemplo de uso
text2 = """
En 1912, el Titanic zarpó con la esperanza y la ilusión de cruzar el Atlántico de una manera nunca antes vista, encapsulando el espíritu optimista de una era que creía fervientemente en el progreso humano y la tecnología.
Esa esperanza se convirtió en desesperación en cuestión de momentos cuando, en una fría noche de abril, un iceberg se interpuso en su camino.
Los sistemas de seguridad, que se consideraban a prueba de fallos, se mostraron trágicamente inadecuados, ya que no había suficientes botes salvavidas para todos los pasajeros.
Las consecuencias fueron devastadoras: más de mil quinientas almas se sumergieron en el helado abrazo del mar, en un suceso que sirvió como un doloroso recordatorio de la vulnerabilidad humana ante las fuerzas de la naturaleza.
"""
bag_of_words = text_to_bag_of_words(text2)
bow2=bag_of_words

for word in bag_of_words.keys():
    print(f"{word}: aparece {bag_of_words[word]} veces en el texto")

# Filtrar palabras con frecuencia mayor que 1
filtered_words = {word: freq for word, freq in bag_of_words.items() if freq > 1}

# Ordenar las palabras por frecuencia
sorted_words = sorted(filtered_words.items(), key=lambda x: x[1], reverse=True)

# Extraer las palabras y sus frecuencias en listas separadas
words, frequencies = zip(*sorted_words)

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies, color='blue', width=1.6)  # Ancho de la barra establecido en 1.6
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.title('Bag-of-Words')
plt.xticks(rotation=90, fontsize=12)  # Hacer que las etiquetas del eje x aparezcan en vertical
plt.yticks(fontsize=10)
for i, freq in enumerate(frequencies):
    plt.text(i, freq, str(freq), ha='center', va='bottom')
plt.show()
```

```python
from gensim.models import Word2Vec
import numpy as np
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

#text1 = "The Titanic sailed with hope and illusion to cross the Atlantic."
#text2 = "The sinking of the Titanic was one of the greatest maritime tragedies."

# Tokenizar los textos
tokens1 = word_tokenize(text1.lower())
tokens2 = word_tokenize(text2.lower())

# Entrenar el modelo Word2Vec
model = Word2Vec([tokens1, tokens2], vector_size=100, window=5, min_count=1, sg=0)

# Calcular el embedding del texto tomando el promedio de todos los embeddings de las palabras
embedding1 = sum(model.wv[word] for word in tokens1) / len(tokens1)
embedding2 = sum(model.wv[word] for word in tokens2) / len(tokens2)

from scipy.spatial import distance

# Calcular la distancia del coseno
cosine_dist = distance.cosine(embedding1, embedding2)

cosine_dist, 1 - cosine_dist
```

```python
from transformers import BertTokenizer, BertModel
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

#text1 = "The Titanic sailed with hope and illusion to cross the Atlantic."
#text2 = text1; #"The sinking of the Titanic was one of the greatest maritime tragedies."

# Tokenizar y calcular embeddings
with torch.no_grad():
    inputs1 = tokenizer(text1, return_tensors="pt", padding=True, truncation=True)
    outputs1 = model(**inputs1)
    embedding1 = outputs1.last_hidden_state.mean(dim=1).flatten()

    inputs2 = tokenizer(text2, return_tensors="pt", padding=True, truncation=True)
    outputs2 = model(**inputs2)
    embedding2 = outputs2.last_hidden_state.mean(dim=1).flatten()

from scipy.spatial import distance

# Calcular la distancia del coseno
cosine_dist = distance.cosine(embedding1, embedding2)

cosine_dist, 1 - cosine_dist
```
