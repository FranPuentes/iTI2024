```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import re

def split(texto, stopwords=[]):
    # Usa una expresión regular para encontrar todas las palabras en el texto.
    palabras = re.findall(r'\w+', texto.lower())
    # Elimina las palabras vacías de la lista.
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stopwords]

    return palabras_filtradas
```

```python
# Oraciones de entrenamiento (cada oración es una lista de palabras)
sentencias = [
    ["gato", "animal", "felino"],
    ["perro", "animal", "canino"],
    ["gato", "maulla"],
    ["perro", "ladra"],
    ["pájaro", "animal", "vuela"],
    ["pez", "animal", "nada"]
]

print(sentencias)
```

```python
# Entrenar el modelo Word2Vec
model = Word2Vec(sentencias, vector_size=10, window=3, min_count=1, workers=1)

# Acceder al vector de una palabra
vector_gato = model.wv['gato']
print(vector_gato);

# Encontrar palabras más similares a "gato"
similar_words = model.wv.most_similar('gato', topn=5)
print("Palabras similares a 'gato':", similar_words)

# Calcular la similitud de coseno entre dos palabras
similarity = model.wv.similarity('gato', 'felino')
print("Similitud entre 'gato' y 'felino':", similarity)

# Realizar operaciones aritméticas con los vectores de palabras
# (gato - felino + canino)
result = model.wv.most_similar(positive=['gato','canino'], negative=['felino'], topn=1)
print("Resultado de gato - felino + canino:", result)
```
